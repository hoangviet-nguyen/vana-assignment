Implementation of task one data exploration
First we load the data into the variables
After everything is initialized we can describe the Data with which we are working
The code to Load the data is in DataLoader.py

```python
import DataLoader
import torch
mnist_trainset = DataLoader.mnist_trainset
mnist_testset = DataLoader.mnist_testset

x_train, y_train = DataLoader.x_train, DataLoader.y_train
x_test, y_test = DataLoader.x_test, DataLoader.y_test

#Desribes the data we are working with
def data_format():
    print("The size of the training dataset is: " + str(len(mnist_trainset)))
    print("The size of the test set is: " + str(len(mnist_testset)))
    print("The size of one image is: " + str(mnist_trainset.data.shape))
    print("The Input container is: " + str(type(mnist_trainset.data[0])))
    print("The output type is: " + str(type(mnist_trainset.targets)))
    print("The data types of the input are: " + str(type(y_train)) + " and " + str(type(x_train)))
    print("Der Input Wertebereich is: ", f"[{mnist_trainset.data.min()}, {mnist_trainset.data.max()}]")

data_format()
```

Here are some sample images with the corresponding labels

```python
import matplotlib.pyplot as plt

labels = [i for i in range(10)]

#Visualizes some sample data with their labels
def visualizeData():
    figure = plt.figure(figsize=(8, 8))
    cols, rows = 3, 3
    for i in range(1, cols * rows + 1):
        index = torch.randint(len(mnist_trainset), size=(1,)).item()
        img, label = mnist_trainset[index]
        figure.add_subplot(rows, cols, i)
        plt.title("labels = " + str(labels[label]))
        plt.axis("off")
        plt.imshow(img.squeeze(), cmap="gray")
    plt.show()

visualizeData()
```

At last we want to check how the spread of the data is. We need an even spread so that we don't run into problems later on

```python
#Initialize the histogram
histogram = [0] * 10

for i in range(len(mnist_trainset)):
    img, label = mnist_trainset[i]
    histogram[label] += 1

#Visualize the spread of the data
def visualizeSpread():
    plt.bar(labels, histogram, edgecolor='black')
    plt.xlabel("Label")
    plt.ylabel("Number of images")
    plt.title("Spread of the data")
    plt.show()

visualizeSpread()


```
